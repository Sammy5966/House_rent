# 5 Bar Charts

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=house_rent_df, x='City')
plt.title('Bar Chart 1: Count of Houses by City')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=house_rent_df, x='City', y='Rent')
plt.title('Bar Chart 2: Average Rent by City')
plt.show()
```
